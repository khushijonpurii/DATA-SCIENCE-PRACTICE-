```python
%%writefile app.py 
#magic command to create app.py file
import streamlit as st
import joblib
model=joblib.load('spam-ham')   #/content/spam-ham 
st.title('SPAM-HAM CLASSIFIER')    #title of web app
ip=st.text_input('enter the message')  #creates a text box in web app
op= model.predict([ip])
if st.button('Predict'):
  st.title(op[0])                  #st.button will create a button with name Predict
  #st.title(op[0])                 #op will be displayed as a title
```
