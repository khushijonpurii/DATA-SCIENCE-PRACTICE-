```python
#before normalisation

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)

x1=np.random.randint(1,50,30)
x1=np.sort(x1)
x2=np.random.randint(10000,70000,30)
plt.plot(x1,x2)
```

```python
#after normalisation
x1min=min(x1)
x1max=max(x1)
x2min=min(x2)
x2max=max(x2)
x1norm=(x1-x1min)/(x1max-x1min)
x2norm=(x2-x2min)/(x2max-x2min)
plt.plot(x1norm,x2norm)
```
