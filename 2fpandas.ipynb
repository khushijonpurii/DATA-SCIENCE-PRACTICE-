```python
import pandas as pd
df= pd.read_csv('https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/laptops.csv',encoding='latin-1')
df
```

```python
df.info()
```

```python
df.shape  #1303 rows and 13 cols
```

```python
df.size
```

```python
df.isnull().sum()
```

```python
#to drop col unnamed:0
df=df.drop(columns='Unnamed: 0')
```

```python
df
```

```python
df.info()
```

```python
#removing the GB from the Ram col and keeping only int values

df['Ram']= df['Ram'].str.replace('GB','')
```

```python
df.info()
```

```python
df
```

```python

#coverting data of Ram col from string to int

df['Ram']= df['Ram'].astype('int64')
df.info()
```

```python
#removing 'kg' from the data of weight col

df['Weight']= df['Weight'].str.replace('kg','')
df['Weight']= df['Weight'].astype('float64')
df.info()
```
