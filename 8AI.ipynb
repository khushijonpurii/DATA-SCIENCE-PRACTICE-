```python
outlook.loc['Sunny','py']
```

```python
temp = pd.crosstab(df['temp'],df['play'])
temp['py'] = temp['Yes']/temp['Yes'].sum()
temp['pn'] = temp['No']/temp['No'].sum()
temp
```

```python
humidity = pd.crosstab(df['humidity'],df['play'])
humidity['py'] = humidity['Yes']/humidity['Yes'].sum()
humidity['pn'] = humidity['No']/humidity['No'].sum()
humidity
```

```python
wind = pd.crosstab(df['wind'],df['play'])
wind['py'] = wind['Yes']/wind['Yes'].sum()
wind['pn'] = wind['No']/wind['No'].sum()
wind
```

```python

# P(Yes | Sunny,Hot,High,Weak) = P(Y) * P(sunny y) * P(Hot Y) * P(High Y) * P(Weak Y)
#ip = ['Sunny','Cool','High','Strong']
ip = ['Rain','Cool','Normal','Weak']
x1 = 'py'
x2 ='pn'

pyes = py * outlook.loc[ip[0],x1] * temp.loc[ip[1],x1] * humidity.loc[ip[2],x1] *wind.loc[ip[3],x1]
pyes
```

```python
pno = pn * outlook.loc[ip[0],x2] * temp.loc[ip[1],x2] * humidity.loc[ip[2],x2] *wind.loc[ip[3],x2]
pno
```

```python
# Normalized Probability

yon = (pyes)/(pyes+pno)
non = (pno)/(pyes+pno)
print(non,yon)
np.argmax([non,yon])
```

```python
#KMean Clustering- unsupervised learning algo
```

```python
# It has unlabeled data (only ip is directly mentioned)
# It computes the centroids and repeats until the optimal centroid is found.

# It assumes how many clusters(k) are there
# Elbow method : Used to find the value of k

# Sum of squared distance between data points and centroid is min
```

```python
# Step 1 : Find the k value using Elbow method
# Step 2: choose k data points in random and assign it to each cluster
# Step 3: Compute the centroid of the cluster
# Step 4: Add all the points one by one and start segregating into various clusters(Distance formula)
# Step 5: Repeat Step 4 till you get optimum value of centroids
```

```python
import pandas as pd
df=pd.read_csv('/content/Mall_Customers.csv')
df
```

```python
df.head()
```

```python
df.shape
```

```python
x=df.iloc[:,3:,].values
```

```python
import matplotlib.pyplot as plt
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
#find k value using elbow method

lendf= len(df)
lendf
```

```python
import numpy as np
sqlendf= np.sqrt(lendf)
sqlendf
```

```python
from sklearn.cluster import KMeans
ssd=[]
for i in range(2,14):
  kmeans = KMeans(n_clusters=i, random_state=0).fit(x)
  ssd.append(kmeans.inertia_)
plt.scatter(range(2,14),ssd)
plt.plot(range(2,14),ssd)
plt.show()
```

```python
#let us assume k=5
```

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(x)
```

```python
kmeans
```

```python
y=kmeans.predict(x)
```
