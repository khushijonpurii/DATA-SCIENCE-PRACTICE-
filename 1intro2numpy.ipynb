```python
#4. MEAN SQUARED ERROR BETWEEN PREDICTED VALUES AND REAL VALUES
''' MSE is the average of the squared error
 that is used as the loss function for least
 squares regression  '''
```

```python
import numpy as np
y_pred=np.array([110,220,330])
y_real=np.array([100,200,300])

#get mse val
#expected op= 466.666
```

```python
error=np.subtract(y_pred,y_real)
error
```

```python
sq_error=np.square(error)
sq_error
```

```python
mean_sq_error= np.mean(sq_error)
mean_sq_error
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_pred,y_real)
```

```python
# Joining 2 lines in a graph

import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1,6)
y = np.square(x)
z = np.power(x,3)

plt.plot(x,y,color='r',ls='--',lw=2,marker = '*',label='Square')
plt.plot(x,z,color='g',ls=':',lw=1,marker = '+',label='Cube')
plt.legend()
plt.show()
```
